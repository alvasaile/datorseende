```python
from ultralytics import YOLO

# Ladda en grundmodell av YOLO26 (vi väljer 'n' för nano, vilket brukar vara snabbast att träna)
model = YOLO('yolo26n.pt') 

# Starta träningen
results = model.train(
    data='./data.yaml', # Uppdatera till din faktiska sökväg
    epochs=100,                        # Börja med 50 epoker för att se hur modellen lär sig
    imgsz=640,                        # Standardstorlek för bilderna
    batch=16,                         # Antal bilder som processas samtidigt
    project='./output',               # Namnet på mappen där resultaten sparas
    device='cuda'                     # Ändra till '0' eller 'cuda' om du har en dedikerad GPU
)
```
