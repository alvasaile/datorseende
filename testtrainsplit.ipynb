```python
import os
import shutil
from sklearn.model_selection import train_test_split

# ==========================================
# 1. DINA SÖKVÄGAR
# ==========================================
source_images = './yolo/datasets/one/images' 
source_labels = './yolo/datasets/one/labels' 
base_dest = 'yolo_dataset' 

# ==========================================
# 2. HÄMTA ALLA FILNAMN
# ==========================================
# Hämta alla bilder och plocka bort filändelsen
all_images = [f for f in os.listdir(source_images) if f.endswith(('.jpg', '.png', '.jpeg'))]
filenames = [os.path.splitext(f)[0] for f in all_images]

# ==========================================
# 3. DELA UPP MED SCIKIT-LEARN (80/10/10)
# ==========================================
# Steg 1: Bryt ut 80% till train och 20% till temp (val + test)
train_files, temp_files = train_test_split(filenames, test_size=0.2, random_state=42)

# Steg 2: Dela temp-filerna på hälften (vilket blir 10% val och 10% test av totalen)
val_files, test_files = train_test_split(temp_files, test_size=0.5, random_state=42)

print(f"Totalt antal filer: {len(filenames)}")
print(f"Train: {len(train_files)} | Val: {len(val_files)} | Test: {len(test_files)}")

# ==========================================
# 4. SKAPA MAPPAR OCH KOPIERA FILER
# ==========================================
splits = {
    'train': train_files,
    'val': val_files,
    'test': test_files
}

for split_name, file_list in splits.items():
    # Skapa mappar om de inte finns
    os.makedirs(os.path.join(base_dest, 'images', split_name), exist_ok=True)
    os.makedirs(os.path.join(base_dest, 'labels', split_name), exist_ok=True)
    
    # Kopiera in filerna
    for filename in file_list:
        # Hitta den exakta bildfilen (för att hantera .jpg, .png etc.)
        img_name = next(img for img in all_images if img.startswith(filename + '.'))
        label_name = filename + '.txt'
        
        # Sökvägar
        src_img = os.path.join(source_images, img_name)
        src_lbl = os.path.join(source_labels, label_name)
        dest_img = os.path.join(base_dest, 'images', split_name, img_name)
        dest_lbl = os.path.join(base_dest, 'labels', split_name, label_name)
        
        # Kopiera bild
        shutil.copy(src_img, dest_img)
        
        # Kopiera textfil (label) om den finns
        if os.path.exists(src_lbl):
            shutil.copy(src_lbl, dest_lbl)
        else:
            print(f"Varning: Hittade ingen label för {img_name}")

print("Datasetet är uppdelat och klart med scikit-learn!")
```
