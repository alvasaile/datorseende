```python

import pandas as pd
from ultralytics import YOLO

# 1. Ladda din tränade modell
model = YOLO('runs\detect\output/train\weights/best.pt') 

print("--- Startar utvärdering på test-datasetet ---")

# 2. Kör utvärderingen på test-datasetet
metrics = model.val(
    data='./data.yaml', 
    split='test',       
    project='./test_output',
    name='test_evaluationtest'
)

# 3. Extrahera övergripande resultat (för alla klasser kombinerat)
results_data = [{
    'Klass': 'Alla klasser',
    'Precision': metrics.box.mp,  # Mean Precision
    'Recall': metrics.box.mr,     # Mean Recall
    'mAP50': metrics.box.map50,   # mAP vid 50% IoU
    'mAP50-95': metrics.box.map   # mAP vid 50-95% IoU
}]

# 4. Extrahera resultat per specifik klass (Dryckeskartong, Konservburk, Pantburk)
# metrics.box.ap_class_index innehåller index för de klasser som fanns med i testet
for i, class_idx in enumerate(metrics.box.ap_class_index):
    class_name = model.names[class_idx]
    results_data.append({
        'Klass': class_name,
        'Precision': metrics.box.p[i],
        'Recall': metrics.box.r[i],
        'mAP50': metrics.box.ap50[i],
        'mAP50-95': metrics.box.ap[i]
    })

# 5. Skapa en pandas DataFrame och spara till CSV
df_results = pd.DataFrame(results_data)

# Spara till en CSV-fil (index=False gör att vi slipper radnummer i filen)
csv_filename = 'test_utvardering_resultat.csv'
df_results.to_csv(csv_filename, index=False, sep=';', decimal=',')

print(f"\nResultaten har sparats i filen: {csv_filename}")
print("Här är en förhandsgranskning av datan:")
print(df_results)
```
